# Código Python para buscar publicaçoes X e Gerar os arquivos .csv no formato normalizado e com label 

```python
import os
import pandas as pd
import tweepy
import tkinter as tk
from tkinter import ttk

# Função para buscar tweets usando a API v2
def search_tweets(bearer_token, query, start_time, end_time, max_results=100):
    client = tweepy.Client(bearer_token=bearer_token)

    query = f"{query} lang:pt"
    tweets = client.search_recent_tweets(query=query, start_time=start_time, end_time=end_time, max_results=max_results, tweet_fields=['created_at', 'public_metrics', 'text', 'author_id'])

    all_tweets = []
    for tweet in tweets.data:
        tweet_data = {
            'ID': tweet.id,
            'AUTHOR': tweet.author_id,
            'URL': f"https://twitter.com/i/web/status/{tweet.id}",
            'DATE': tweet.created_at,
            'LIKES': tweet.public_metrics['like_count'],
            'COMMENTS': tweet.public_metrics['reply_count'], 
            'CONTENT': tweet.text,
            'SERVICE': 'Twitter',
            'SENTIMENT': 'NEUTRAL',  # Valor padrão, ajustar conforme necessário
            'ENGAGEMENT': tweet.public_metrics['like_count'] + tweet.public_metrics['reply_count'],  # Exemplo de cálculo de ENGAGEMENT
            'SHARES': tweet.public_metrics['retweet_count']  # 'SHARES' para Twitter é 'retweet_count'
        }
        
        all_tweets.append(tweet_data)
    
    return all_tweets

# Função para organizar os dados dos tweets
def organize_tweet_data(tweets):
    df = pd.DataFrame(tweets)
    df.columns = df.columns.str.upper()
    return df

# Função para converter sentimento
def convert_sentiment(sentiment_text):
    sentiment_map = {'NEUTRAL': 0, 'POSITIVE': 1, 'NEGATIVE': 2}
    return sentiment_map.get(sentiment_text.upper(), 0)

# Função para salvar os dados em arquivos CSV individuais
def save_tweets_to_csv(organized_tweets, output_directory):
    for index, tweet in organized_tweets.iterrows():
        tweet_id = tweet['ID']
        service = tweet['SERVICE']
        date = pd.to_datetime(tweet['DATE']).strftime('%Y%m%d%H%M')
        sentiment = convert_sentiment(tweet['SENTIMENT'])
        index_formatted = str(index).zfill(7)
    
        filename = f"{service}_{tweet_id}_{date}_{sentiment}_{index_formatted}.csv"
        filepath = os.path.join(output_directory, filename)
    
        df_single_row = pd.DataFrame([tweet])
        df_single_row.to_csv(filepath, index=False)

# Função para rodar o código a partir do formulário
def run_code(bearer_token, query, start_time, end_time):
    tweets = search_tweets(bearer_token, query, start_time, end_time)
    organized_tweets = organize_tweet_data(tweets)
    save_tweets_to_csv(organized_tweets, output_directory)
    print("Dados salvos com sucesso.")

# Configuração do formulário
def create_form():
    def on_submit():
        bearer_token = bearer_token_entry.get()
        query = query_entry.get()
        start_time = start_time_entry.get()
        end_time = end_time_entry.get()
        root.destroy()
        run_code(bearer_token, query, start_time, end_time)

    root = tk.Tk()
    root.title("Twitter Data Fetcher")

    frame = ttk.Frame(root, padding="10")
    frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

    bearer_token_label = ttk.Label(frame, text="Bearer Token:")
    bearer_token_label.grid(row=0, column=0, sticky=tk.W)
    bearer_token_entry = ttk.Entry(frame, width=30)
    bearer_token_entry.grid(row=0, column=1, sticky=(tk.W, tk.E))

    query_label = ttk.Label(frame, text="Query:")
    query_label.grid(row=1, column=0, sticky=tk.W)
    query_entry = ttk.Entry(frame, width=30)
    query_entry.grid(row=1, column=1, sticky=(tk.W, tk.E))

    start_time_label = ttk.Label(frame, text="Start Time (YYYY-MM-DDTHH:MM:SSZ):")
    start_time_label.grid(row=2, column=0, sticky=tk.W)
    start_time_entry = ttk.Entry(frame, width=30)
    start_time_entry.grid(row=2, column=1, sticky=(tk.W, tk.E))

    end_time_label = ttk.Label(frame, text="End Time (YYYY-MM-DDTHH:MM:SSZ):")
    end_time_label.grid(row=3, column=0, sticky=tk.W)
    end_time_entry = ttk.Entry(frame, width=30)
    end_time_entry.grid(row=3, column=1, sticky=(tk.W, tk.E))

    submit_button = ttk.Button(frame, text="Submit", command=on_submit)
    submit_button.grid(row=4, column=0, columnspan=2)

    root.mainloop()

# Definir o diretório de saída
output_directory = r'C:\madonna\split'

# Criar e exibir o formulário
create_form()
```
